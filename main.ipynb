```python
import json
```

```python
years=['2020','2021','2022','2023','2024']
datas=['1.json','2.json','3.json','4.json']
```

```python
collected_data=[]
for year in years:
    for data in datas:
        try:
            with open(f'/Users/bootlabs/guvi project-1/pulse/data/aggregated/insurance/country/india/{year}/{data}','r') as file :
                data=json.load(file)
                collected_data.append(data)
        except Exception as error:
            print(f'error due to {error}')            
```

```python
import pandas as pd 
pd.json_normalize(collected_data)
# collected_data
```
